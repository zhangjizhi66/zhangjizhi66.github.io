# MNT 实验背景

## 实验背景

### 核反应的过程与分类

<img src="img/Nuclear_reaction.jpeg" style="zoom:40%">

- 独立粒子阶段
   - 卢瑟福散射、库仑散射或库仑激发
      - 原子核之间没有直接接触，仅有库仑场发挥作用
      - 弹靶的碰撞参量较大
   - 形状弹性散射 shape elastic scattering
      - 独立粒子阶段形成的弹性散射

- 复合系统阶段
   - 直接作用 direct reaction
      - 弹核较轻，靶核较重
      - 弹核将能量交给靶核的一个或多个核子，或引起靶核的集体激发，随后基本完整飞出
      - 时间尺度快，约$10^{- 22}s$，相当于具有费米动量的核子穿越整个原子核的时间
      - 在质心系中不是90度对称的
   - 双核子系统 di- nuclear system
      - 弹靶接触后形成di- nuclear system，两者之间存在一个neck
      - 生存时间短，约$10^{- 21}- 10^{- 20}s$，因此仅能达到部分热平衡
      - 一般在弹靶均较重的时候形成
   - 双核子系统的分裂 quasi- fission / deep inelastic transfer / strongly damped collisions
      - 双核子系统没有形成复合核，在达到热平衡之前重新分离（平衡前发射）
      - 弹靶分离后大致保持原来的特征，称为类弹和类靶
      - 在参与碰撞的原子核之间进行质子中子的快速分配，时间尺度约$10^{- 22}s$，分配之后的状态称为N/Z平衡
      - 兼具直接作用与复合核的特点
      - - 在质心系中不是90度对称的，多数产物向前飞出（也有部分体现出类轨道行为orbiting- like behavior，向负角度发射）
      - - 产物的动能与出射道的Coulomb barrier接近，与束流能量无关
   - 复合核
      - 在直接作用或双核子系统的基础上，弹靶进一步熔合形成复合核
      - 生存时间长，约$10^{- 19}- 10^{- 16}s$，可以达到热平衡
      - 复合核的性质与弹靶种类及其形成过程都没有关系
      - 产物的动能与出射道的库仑势垒接近，与束流能量无关
      - 退激方式（能量由低到高）
      - - 熔合蒸发反应：1- 10MeV/A，可以布居高自旋态
      - - 熔合裂变：复合核的裂变
      - - 散裂反应：大于40MeV/A，发射若干块小的核子集团
      - - 碎裂反应：核子全部被打散
   - 复合核弹性散射 compound elastic scattering
      - 复合核发射一个与弹核完全相同（结构、能量）的粒子，即复合核阶段形成的弹性散射

### 多核子转移
- 发生在双核子系统阶段
- 根据核子转移数目、能量耗散强度等进行分类
   - 准弹性散射 quasi elastic scattering （也可能在双核子系统形成之前，或许没有明确的界限）
   - 深度非弹性散射 deep inelastic scattering
   - 准裂变 quasi- fission

### 重核两体反应
- 准弹性散射
   - 碰撞参量大，产物与弹靶相同或相差几个核子，同时被轻微地激发
   - 细节包括库仑激发、非弹性散射以及少量核子的转移
   - 布居的态较低，观测到的gamma多重性低

- 熔合裂变
   - 碰撞参量小，形成复合核；随后复合核在很高的角动量下发生裂变
   - 碎片分布以对称裂变的产物为中心，分布较广
   - 碎片被强烈地激发，随后经历粒子发射和高多重性的gamma退激

- 深度非弹性散射
   - 中等碰撞参量，大范围的质量、能量、角动量转移

### 库仑势垒与反应势垒
- 库仑势垒$$V_c=\frac{Z_pZ_te^2}{R_C}$$, $$R_C=r_{0}(A_p^{1/3}+A_t^{1/3})$$
- 反应势垒
   - 原子核半径：电荷密度下降到中心密度的一半处，$$R_i=1.12A_i^{1/3}- 0.94A_i^{- 1/3}$$，$$i=projectile/target$$
   - 反应中的有效核间距：由实验给出，$$R_{int}=R_p+R_t+3 fm$$
   - 反应势垒：$$B_{int}=\frac{Z_pZ_te^2}{R_{int}}$$

### 两体动力学
- 生成核的动量由动量守恒给出：$$P_{1,2}=\frac{P_0 \sin{\theta_{2,1}}}{\sin(\theta_1+\theta_2)}$$，其中$P_0$为束流动量，$P_{1,2}$为核子动量，$\theta_{1,2}$为核子散射角

### 擦边角
- 核反应截面在擦边角(grazing angle)处最大，小于（对TLF为大于）擦边角则仅发生库仑散射或卢瑟福散射
- 碰撞参数d与散射角&theta;的关系为$$d=\frac{Z_tZ_pe^2}{4\pi\epsilon_0E_k}(1+\csc\frac{\theta}{2})$$，其中$Z_p,Z_t$为原子序数，$E_k$为动能
- 当碰撞参数与两核子半径之和相等时认为它们发生接触，则$$d=1.2(A_t^{\frac13}+A_p^{\frac13})\ fm$$，由此可以定出两特定核子的擦边角

## 理论模型

### GRAZING

- 用于计算反应截面
- 将弹靶描述为在球形附近振动的独立核子
- 只考虑在擦边角附近的反应，更小碰撞参量下的 quasi- fission 等反应没有考虑
- 没有考虑产物的裂变，改进代码为GRAZING- F
   - 对于136Xe+208Pb @ Ecm=450MeV，GRAZING和GRAZING- F的结果相同，表明裂变不是其产物主要的退激方式
- 形状因子
   - 用于描述单个核子在弹靶之间转移的行为
   - 调节中子的形状因子：改变截面- N单峰曲线的宽度
   - 调节质子的形状因子：改变截面- N单峰曲线的峰值

### Time Dependent Hartree- Fock

- 弹靶考虑为球形；核子是独立的，处在单粒子轨道上
- 与GRAZING相比。考虑了深度非弹以及准裂变道
- 计算得到质子和中子分别出现在PLF和TLF中的概率，进而得到截面
   - N个中子和Z个质子出现在PLF中的概率（碰撞参量为b）：$$P_{Z,N}(b)=P_Z(b)P_N(b)$$
   - 反应截面：$$\sigma(Z,N)=\int_{b_{min}}^{\infty}2\pi{b} P_{Z,N}(b)db$$，$b_{min}$为不发生弹靶熔合的最小碰撞参量
- TDHF与GRAZING的结果较为符合；但在质子转移数较多时，两者与实验结果均偏离较大，没有很好的解释（2015）

### Langevin (by Zagrebaev and Greiner)

- 强调了壳效应，认为闭壳处的势阱是PLF/TLF形成的驱动力
- 反应过程中系统势能与核间距、形变以及弹靶的质量差异有关；因此这些变量的演化遵循能量最低原理
- 描述系统的演化采用朗之万形式的方程
   - $$\frac{d\eta_Z}{dt}=\frac{2}{Z_{CN}}D_Z^1+\frac{2}{D_{CN}}\sqrt{D_Z^2}\Gamma(t)$$
   - $$\frac{d\eta_N}{dt}=\frac{2}{N_{CN}}D_N^1+\frac{2}{D_{CN}}\sqrt{D_N^2}\Gamma(t)$$
      - $Z_{CN},N_{CN}$为复合核的核子数目
      - $D^1,D^2$为输运系数
      - $\eta_Z,\eta_N$为质子、中子的质量偏移：$\eta_Z=\frac{2Z- Z_{CN}}{Z_{CN}}$，Z为一个碎片的质子数
- 模型假设多核子转移是一系列单核子转移的组合，则输运系数可以表示为跃迁概率的线性组合
   - $$D_{Z,N}^1=\lambda_{Z,N}^{(+)}(A\rightarrow{A}+1)- \lambda_{Z,N}^{(- )}(A\rightarrow{A}- 1)$$
   - $$D_{Z,N}^2=\frac12[\lambda_{Z,N}^{(+)}(A\rightarrow{A}+1)+\lambda_{Z,N}^{(- )}(A\rightarrow{A}- 1)]$$
      - 跃迁概率由能级密度给出：$$\lambda_{Z,N}^{(\pm)}=\lambda_0\sqrt{\frac{\rho(A\pm{1})}{\rho{A}}}$$，其中能级密度$\rho\sim{e}^{\sqrt{aE^- }}$，a为常数
- primary product截面公式
   - $$\frac{d^2\sigma_{Z,N}}{d\Omega{dE}}(E,\theta)=\int_0^\infty{bdb}\frac{\Delta{N}_{Z,N}(b,E,\theta)}{N_{tot}(b)}\frac1{\sin\theta\Delta\theta\Delta{E}}$$
      - $\Delta{N}_{Z,N}(b,E,\theta)$：在碰撞参量b下生成$Z,N,E+\Delta{E},\theta+\Delta\theta$的核的数目，其中$\theta$为质心系下的坐标
- primary product的退激通过统计模型计算
- 模型在描述纯质子转移时，与实验结果符合较好；描述纯中子转移时截面预测稍偏高
- 模型将核子转移分为“对称化”和“非对称化”
   - 48Ca+248Cm：对称化，248Cm倾向于208Pb
   - 238U+248Cm：非对称化，238U倾向于208Pb

### Di- Nuclear System model

- 将核子转移看作扩散过程
- 计算单个核子的转移概率，再将其求和
- 认为激发能按照质量比例在primary fragments之间分配
