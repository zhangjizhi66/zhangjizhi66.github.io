# Krolas: Gamma coincidence study of 208Pb+350MeV 64Ni collisions

[KrolasNPA2003](../pdf/1-s2.0-S0375947403015446-main)


### 实验方法

- 208Pb, 30 mg/cm2, 厚靶
- 64Ni @ 350MeV
- 探测阵列：OSIRIS, 11 Compton suppressed Ge detectors, 48 BGO elements
- 脉冲周期: 69 ns
- 在束时间: 84 h

### 数据分析

#### 在束数据
- 深度非弹产物
   - population is mainly yrast @ lowest excitation region
   - 因此，最低的几条yrast transitions的符合强度代表了大部分的产出
   - 特别地，对于大质量转移（10以上）的偶偶核产物，4+到2+到0+的跃迁代表了90%以上的产物总截面
- 准弹散产物
   - 需要判断几条跃迁的符合强度在总截面中的比例，才能计算总截面

#### 熔合反应
- 由于束流能量350MeV仅仅比库仑势垒高出12%，因此熔合反应的截面是低的
   - 在更高能量380MeV下，复合核更容易形成，熔合截面变高
- 部分熔合裂变产物被探测到，位于A～140区域

#### 截面计算
- 截面结果的准确度大约是20%

#### 中子发射
- 认为大多数产物是稳定的或丰中子的，因此初始产物只经历中子发射过程，其他过程可以忽略
- 在弹靶附近，中子发射的数目与初始产物的中子数相关性较强；在大质量转移区则相关性较弱
- 通过分析中子发射数目，可以还原出初始产物截面。主要有交叉符合以及缺失质量两种分析方法。
   - 结果表明，类靶碎片平均发射不超过6个中子
   - 两种方法算出的中子发射数符合较好，差异不超过0.4个中子

##### 1、交叉符合分析法 cross-coincidence analysis
- 使用在束数据
- 用一种产物（如Ni）中最强的gamma开窗，计算partner（Z值互补，如Pb）的同位素产出分布
- 进而以产出为权重，计算同位素质量数的平均值，从而算出平均中子发射数
- 除Ni-Pb之外，其他情况只对重核开窗，从而得到中子发射数与重核同位素的关联关系。相比轻核开窗，结果更完整可靠。
- 注意：这样的分析过程并不能分辨两个碎片各自发射了多少中子
   - 需要假设激发能在碎片之间按质量比例分布

##### 2、缺失质量分析法 missing mass analysis
- 对于截面很低的产物，一般缺乏其同位素的截面结果，无法得到同位素产出分布时，用本方法
- 将Z值互补的两种产物的质量数按产出权重相加 (yield weighted mass)，在弹靶总质量272中减之，得到总中子发射数，再在碎片中分配中子发射数
- 该方法对于大质量转移区较为适用，这一区域中子发射数与同位素关系不大

#### N/Z平衡

<img src="../img/N-Z_ni64_pb208.jpeg" style="zoom:50%">

- 弹核64Ni是丰中子核，但在反应中容易吸收额外的中子；产物中包含68Ni，69Ni
   - 但在大质量转移时产物分布回到了稳定区，这个现象与N/Z平衡有关

- N/Z平衡指的是反应之后，初始碎片的N/Z比值会向复合核的N/Z比值靠近
- 将质量数相同的所有产物的N/Z值取平均，并作出右图
   - 使用初始产物的N/Z值，并讨论了中子发射数不同的分配方式
   - 图像表明，历经很小质量的转移后（～3u），碎片的N/Z就会迅速变化为几乎固定的值；更多的质量转移不会明显改变这一数值

- N/Z平衡理论模型 by Swiatecki
   - 考虑两个相互接触的球形核，它们在液滴模型下总能量最低时，达到N/Z平衡；这种考虑高估了N/Z平衡所能进行的程度
   - 改进：考虑处于分裂点 scission point 的两个有形变的核；但模型中没有直接考虑形变，而是认为通过拟合球形核之间的间距，可以粗糙地估计形变程度
   - 通过调节两球形核中心的距离参数，来改变总能量中占主要贡献的库仑能部分，从而达到与实验数据的最佳拟合
      - 结果表明，两球形核表面之间必须加入3-10fm不等的距离才能大致复现实验数据，说明反应中原子核经历了相当程度的形变
      - 若中子发射数平均分配，拟合出的球形核间距更小（相比按质量分配的情况），意味着核的形变更小

- 另一种模型
   - 为了克服液滴模型不够精确的特点，将总能量取为质量损失与库仑能的和。其中质量损失采用实际的原子核质量来计算
   - 与实验结果的符合程度并无明显改善；拟合出的核间距与Swiatecki模型较为符合

- 在另外两篇文献中，作出了N/Z比值与总能量损失的关联；本文中产物截面是所有反应类型之和，无法作出这种关联

- 为了更精确地研究N/Z平衡的过程，作者建议使用更加乏中子的弹核58Ni，从而增加弹靶的N/Z的差异

##### 结论
- N/Z平衡说明，如果用N/Z值较小的丰中子弹核击打N/Z值较大的丰中子靶核，则靶核依然可以向弹核转移中子，从而形成更极端的丰中子核
   - 在大质量转移时，中子发射过程增强，产物多在稳定线附近
