==实验方法==

===gamma射线探测===

*gamma射线与物质的作用截面
**光电效应：<math>\sigma_{pe}=k_{pe}\frac{Z^{4.5}}{E^3}</math>，其中k为常数，Z为探测材料原子序数，E为gamma能量，这表明高Z值的材料更容易发生光电效应；100keV以下的gamma与半导体材料主要发生光电效应
**康普顿效应：<math>\sigma_{cs}=k_{cs}\frac{Z}{E}</math>，当gamma能量提高时主要发生康普顿效应；康普顿散射光子一般能量较低，有足够概率发生光电效应，因此这光电子与康普顿电子有一定概率会加合成全能峰，在0.2-2MeV的范围内这是全能峰的主要来源；如果光电子逃逸，则形成康普顿平台，这时需要用高Z值高密度的BGO（锗酸铋）作反康
**电子对产生：<math>\sigma_{pp}=k_{pp}{Z^2}(\ln{E}-\ln(1.022\ MeV))</math>，是高能gamma的主要反应。在能谱中产生3个峰：电子对动能沉积、湮灭光子全部沉积——全能峰；电子对动能沉积、1个湮灭光子逃逸——全能峰-0.511MeV；电子对动能沉积、2个湮灭光子逃逸——全能峰-1.022MeV

*gamma射线的吸收：<math>N=N_0(1-e^{-\mu{x}})</math>，其中线性吸收系数<math>\mu=n\sigma</math>，n为原子数密度；平均自由程<math>\lambda=\frac1{\mu}</math>

*gamma射线在Ge材料中的沉积：1MeV的电子在Ge中沉积的能量约有5%是韧致辐射；考虑电离损失和韧致辐射后，计算得电子在Ge中的射程约几个mm（对于感兴趣的能量范围）

*gamma探测器效率受限因素：Ge晶体间的BGO占据了相当一部分立体角

==互补碎片研究 study of complementary fragments==

*与交叉符合分析法的区别
**多普勒修正使得我们能够鉴别prompt gamma来自于哪个碎片，而交叉符合分析法被用于厚靶实验，不能作这种鉴别

*用含有isomer的碎片的delayed gamma开窗，看另一碎片的prompt gamma

*delayed gamma应当不包含来自其他核的杂质成分
**可以通过观察开窗谱中是否含有除互补核之外其他核的特征X射线，来判断是否有Z等于其他值的杂质核（例如用134Xe开窗，开窗谱中应仅有Pt的特征X射线）

*开窗谱中gamma的强度不能用来估计中子发射数目（即使用探测效率和内转换系数作了修正）；这是因为一些gamma会卡在isomer上，从而降低探测到的gamma强度

*如果用这种方法研究Pt，那么198Pt的gamma还可能来自库仑激发；另外，192Pt和198Pt含有isomer，因此平均中子发射数求不出来

==事件多重性与角动量转移==

[[File:fold_spin.jpg|thumb|400px]]
[[File:fold_angle.jpg|thumb|700px]]

*矩阵：延迟gamma——散射角——多重性
**延迟gamma用于挑选产物

*分别用198Pt的8+、6+和2+态向下的跃迁开窗，看多重性分布
**高自旋的态下，平均看到的多重性稍高，且开始出现多重性非常高的事件
**低多重性的事件对应于库仑散射；高多重性的事件对应于非弹性过程

*多重性的角分布
**低多重性的事件出现在类弹的前角和类靶的大角度，对应于库仑激发（库仑激发产生低多重性事件）

*讨论
**gamma多重性代表了相互作用的非弹性程度
**对于擦边角附近的准弹性散射，转移的角动量少，事件多重性较低；这一区域可以运用角动量的滚动模型 rolling model
**对于深度非弹性散射，两体间的作用更为深入，接触更牢固。如果接触时间足够长，就可以运用角动量的粘连模型 sticking model

*角动量模型
**角动量守恒给出：<math>L=\mu(r_p+r_t)^2\omega+I_p\omega_p+I_t\omega_t</math>，其中<math>L</math>为初始角动量，<math>r_p,r_t</math>为核子半径，<math>I_p,I_t</math>为核子转动惯量，<math>J_{p,t}=I_{p,t}\omega_{p,t}</math>为核子内禀角动量，<math>\mu</math>为约化质量，<math>\omega</math>为两核子相对角速度
**滑动模型(sliding model)：<math>J_p=J_t=0</math>，即角动量完全没有转移到核子内部
**粘连模型(sticking model)：<math>\omega=\omega_p=\omega_t</math>，即两核子一起旋转，相当于粘在了一起
**滚动模型(rolling model)：<math>r_p(\omega_p-\omega)=r_t(\omega_t-\omega)</math>，即两核子的接触点相对系统质心保持静止，是前两个模型的中间情况；再假设两核子从刚接触到动力学平衡的过程中，接触点存在摩擦力<math>F</math>，而核子的内禀角动量全部来源于<math>F</math>的贡献，则有<math>\int{F}dt=\frac{J_p}{r_p}=\frac{J_t}{r_t}</math>，即<math>\frac{\omega_p}{\omega_t}=\frac{I_tr_p}{I_pr_t}</math>
