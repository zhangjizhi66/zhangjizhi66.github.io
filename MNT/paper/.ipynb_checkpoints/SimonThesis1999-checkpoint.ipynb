# Simon: Spectroscopy of Neutron-Rich Nuclei Produced in the Spontaneous Fission of 252Cf

[SimonThesis1999](../pdf/Simonthesis.pdf)

### 概述

- 薄靶：符合测量——退激gamma射线、散射与反冲核；多普勒展宽可作修正

- 厚靶：能级半衰期与阻停时间可比拟的时候，gamma谱出现多普勒展宽

- 通过测量两生成核的TOF差值及分离角，重建反应动力学，辨认TLF（类靶）和PLF（类弹），进而逐事件作多普勒修正，分辨gamma的来源（PLF或TLF）

- particle-gamma符合测量方法
   - 库仑激发实验：在不同激发强度下测量gamma，获得激发能级的B(E&gamma;)值
   - 核子散射角的信息：用来选择擦边角（grazing angle，弹性散射与非弹性散射的分界）附近的反应，加强核子转移的反应道
   - gamma信息：确定反应产物

- CHICO：Compact Heavy Ion COunter，University of Rochester

- 实验：薄靶实验，CHICO与GammaSphere探测阵列

### 实验装置

#### GammaSphere

<img src="../img/Gammasphere_sketch.png" style="zoom:100%">

- 110个HPGe，直径7.1cm，长8cm，距球心25cm；单个HPGe探测效率20%，分辨2.5keV@1.33MeV（0.188%）；全阵列覆盖4&pi;立体角的47%

- Ge晶体周围包裹BGO，用于减康普顿本底，将峰总比从0.25提高至0.6@1.33MeV

- 单个HPGe对球心张角仅15°，降低了加合峰计数，也减小了多普勒展宽

- 总的峰效率达10%，对多重符合事件(high-fold events)也能达到较高的统计，进而通过“拆对”提高lower-fold event的统计

- 与CHICO耦合后，GammaSphere的峰效率以及峰总比均下降约12%

#### CHICO

- 设计要求：收集足够的核子信息，以重建反应动力学；CHICO主要用来探测两体反应
   - 散射角：核子出射轨迹与束流的夹角，若以束流方向为极轴则为极角。对于两体动力学，由散射角可以定出核子动量：<math>P_{1,2}=\frac{P_0 \sin{\theta_{2,1}}}{\sin(\theta_1+\theta_2)}</math>，其中<math>P_0</math>为束流动量，<math>P_{1,2}</math>为核子动量，<math>\theta_{1,2}</math>为核子散射角
   - 飞行时间：为了区分以上定出的两个核子动量的解，需要飞行时间的信息。CHICO中从靶到探测器的最短飞行距离为12.8cm，在库仑散射实验中，飞行时间差大约为ns量级，因此探测器的时间分辨必须为亚纳秒

- 根据设计要求，探测器选用PPAC(Parallel Plate Avalanche Counter)
   - 具有良好的角分辨及300ps左右的时间分辨
   - 对粒子的探测率达<math>10^5/s</math>
   - 防辐射损伤；探测阈值可调节；结构简单，质量轻
   - 探测材料低Z值，对gamma的探测几乎没有影响

- CHICO的具体结构与参数
   - CHICO的对称性应与GammaSphere一致，这样可以将关于某一核子散射角对称的HPGe测得的谱相加，减少需要分析的谱的数量。GammaSphere是五角对称的，而出射的核子多数是两体，因此CHICO采用十角对称的布局：20片梯形PPAC，分成两个全同的装配体，耦合在与束流垂直的平面上(<math>\theta=90^\circ</math>)。
   - PPAC探测的核子散射角（极角&theta;）范围为<math>12^\circ-85^\circ</math>；而单片PPAC探测的方位角&phi;范围为<math>28^\circ</math>，相邻PPAC间的探测盲区为<math>8^\circ</math>。CHICO整体能够覆盖4&pi;立体角的69%。
   - PPAC的法线与束流的夹角为<math>49^\circ</math>；最短飞行距离12.8cm
   - 外壳：直径35.6cm的铝壳，厚度1.58mm
   - PPAC窗膜：0.9&mu;m厚mylar膜(<math>120\mu{g}/cm^2</math>)
   - 钨准直孔：入口直径6.4mm，出口直径5.1mm，中间放置直径3.2mm的金箔（距靶10.2cm）
   - 靶盘可放置4片样品，并可在不打开靶室的情况下，旋转靶盘改变样品类型

#### PPAC

[[File:PPAC阳极侧面.png|thumb|400px]]

- 阳极：<math>100\mu{g}/cm^2</math>厚的聚丙烯，加上<math>20\mu{g}/cm^2</math>厚的铝涂层，在方位角&phi;方向分割为互相绝缘的两部分，面积分别占1/3和2/3

- 两体动力学中出射核子一定共面（忽略轻粒子的发射），且由于CHICO的对称性，使得方位角&phi;的分辨率由阳极较小的部分所覆盖的方位角决定，约为9.3°；这样在保证分辨率不变的情况下，减少了电子学的通道数量

- 阳极和阴极信号由带状传输线传到放大器，带状传输线包含4条50欧姆阻抗的电缆(trace)，2条接阴极，另2条接阳极并加约400V的偏压

- 阳极和阴极之间由3.2mm厚的垫片隔开；垫片预留通气孔，使PPAC中的气体能够循环流动，避免气体受辐照损伤导致探测器性能下降

- 阴极包含3层：有效层、接地层和背面层
   - 有效层（下图左侧）按极角&theta;间隔1°分割为宽度不均匀的条，每条经过PPAC边缘的镀通孔(plated-thru holes)分别连接到背面层（下图右侧）的1个tap上，相邻tap的延迟为1ns；通过信号到达延迟线两端的时间差可以定出入射条，进而确定极角&theta;
   - 接地层(ground plane)用于将有效层和背面层进行电容隔离；相比无接地层的情形，输出信号的幅度和上升时间都更优

### 实验方法

#### 位置测量与刻度

- 如上节所述，PPAC阳极以9.3°的分辨测量方位角&phi;，阴极以1°的分辨测量极角&theta;

- 下图左侧是CHICO对反应86Kr+63Cu@265MeV测量的位置谱
   - 图中每个峰半宽约1°，表明散射角（极角&theta;）的分辨率为1°
   - &theta;=60°处的谷是由于该处的阴极条被窗膜的支架代替，宽度为1°；在&theta;=25°处也出现了谷，是由于63Cu在&theta;=60方向被支架挡住，相应的86Kr飞向&theta;=25°，不满足符合条件，没有被探测；而飞向&theta;=25°的63Cu仍能被探测到
   - &theta;<20°的小角度散射被mask挡住，因为这一区域的计数率很高；相应地&theta;>67°的散射由于符合条件的原因也没有被探测

- 位置刻度
   - &theta;=60°处的谷可以用来作粗略的刻度，更精确的刻度需要用到弹性散射动力学关系
   - 从两体动力学可以算出出射核子的（极）角关联关系。于是可以将任意两片相对的PPAC测得的极角作图。这时，PLF入射PPAC1且TLF入射PPAC2，以及PLF入射PPAC2且TLF入射PPAC1，这两种情况的代表点分属两条不同的曲线。
   - 这两条曲线关于y=x对称，且其交点满足<math>\cos\theta=\frac12\sqrt{1+\cfrac{m_p}{m_t}}</math>，其中<math>m_p</math>为弹的质量，<math>m_t</math>为靶的质量
   - 对于CHICO的前半部分，如下图右侧所示，采用两套弹靶系统，每对PPAC获得2个刻度点，定出刻度系数
   - 用来刻度的反应必须是弹性散射，这样才能使用能量守恒，得出确定的核子角关联曲线
   - 对于后半部分，使用252Cf产生自发裂变，裂变产物共线飞出，从而利用前半部分的刻度数据对后半部分进行刻度

<img src="../img/PPAC能量刻度.png" style="zoom:90%">

#### 时间测量与粒子鉴别

<img src="../img/CHICO粒子鉴别.png" style="zoom:90%">

- 将两核子的时间差与它们各自的散射角作图，可以进行粒子鉴别，如右图。其中计数最高的区域位于擦边角附近

- 核子质量公式<math>m_{1,2}=\frac{\Delta{TOF}+\cfrac{d_{2,1}}{P_{2,1}}(m_p+m_t)}{\cfrac{d_1}{P_1}+\cfrac{d_2}{P_2}}</math>，其中<math>d,P</math>分别是核子的飞行距离和动量；动量由散射角算出

- 从以上公式可以给出核子的质量谱，分辨约5%

- TOF差值的测量分辨为0.7ns，影响因素如下
   - 探测器本征分辨
   - 核子在靶材料中的能量岐离（约0.7MeV）和角度岐离（约0.2°）
   - 靶的厚度
   - 束斑宽度（3mm），对分辨的贡献最大，因为总的飞行距离仅13cm
   - 电子学分辨（用一个信号给自己开门的方法来测量）
