# 磁谱仪设计


### 目标

- 将能量连续的241Am源用磁场色散，产生1-3MeV的单色alpha粒子

- 设磁谱仪的接收立体角为&Omega;(msr)，分辨率为&eta;(%)；放射源活度20kBq，能量为5.486MeV以下均匀分布，则1MeV的峰计数率（每小时）

$$N=\frac{\Omega\ msr}{4\pi\ sr}\times20kBq\times\frac{\eta\%\times1\ MeV}{5.486\ MeV}\times3600\approx10\Omega\cdot\eta$$

- 要求：1MeV峰计数率20/小时，分辨率0.4%

### 模型

<img src="img/Parameters_of_miniature_magnetic_spectrometer.png" style="zoom:25%">

- 放射源置于点P，磁场位于圆弧$Q_1Q_2,Q_1^\prime Q_2^\prime$之间，方向垂直纸面向里。束流的中心轨道为$PQ_1Q_1^\prime P_1^\prime$。当粒子出射时的偏移角为$\mathrm{\Delta}\varepsilon_1$时，轨道为$PQ_2Q_2^\prime P_2^\prime$。像平面为$P_1^\prime P_2^\prime$。

- $\phi$为中心轨道在磁场中的偏转角，r为轨道半径。$l_1,l_2$分别为入口和出口的漂移距离，即物距和像距。当物点P和像点$P_1^\prime$在磁场外部时，取正值。

- $\varepsilon_1,\varepsilon_2$分别为入口和出口处的倾斜角，即束流的入射角和出射角。当中心轨道的圆心$O_1$与放射源P位于圆弧$Q_1Q_2$的过$Q_1$点的法线的同一侧时，$\varepsilon_1$取正值。$\varepsilon_2$同理。

- $R_1,R_2$分别为入口和出口处磁场边界的曲率半径，即圆弧$Q_1Q_2,Q_1^\prime Q_2^\prime$的半径。当曲率中心与磁场位于磁场边界的同一侧时，$R_1,R_2$取正值。

- $\mathrm{\Delta}\varepsilon_1$为某一粒子从放射源中出射时相对中心轨道的偏移角，偏向远离$O_1$的一侧时取正值；$\mathrm{\Delta B}$为粒子在过像点$P_1^\prime$垂直于中心轨道的平面（即像平面）上的水平像差。

### 一阶效应（xz平面）

- 将束流线上xz平面方向各部分（物距漂移、入射角、二极磁铁、出射角、像距漂移）的R矩阵（参考[束流光学](particle_optics.html)）倒序相乘，得到系统的R矩阵。成像条件要求矩阵R的第1行第2列元素为0，这样就给出了1个约束方程。我们用它来定出像距：

$$l_2(\phi,l_1,\varepsilon_1,\varepsilon_2)=-r\frac
{\bigg(\cfrac{r}{l_1}+\tan\varepsilon_1\bigg)\tan\phi+1}
{\bigg[\bigg(\cfrac{r}{l_1}+\tan\varepsilon_1\bigg)\tan\phi+1\bigg]
\tan\varepsilon_2+\cfrac{r}{l_1}+\tan\varepsilon_1-\tan\phi}$$

- R的第1行第3列元素为单位动量分散导致的像在水平方向的位移，称为色散：

$$D=\frac{\Delta{x}}{\delta{p}/p}=r\bigg[(1-\cos\phi)+\frac{l_2}{r}(\sin\phi+(1-\cos\phi)\tan\varepsilon_2)\bigg]$$

- R的第1行第1列元素为像的放大率：

$$M_x=\frac{l_2}{r}\big[\sin\phi(\tan\varepsilon_1\tan\varepsilon_2-1)+\cos\phi(\tan\varepsilon_1+\tan\varepsilon_2)\big]+\cos\phi+\sin\phi\tan\varepsilon_1$$

### 分辨率

<img src="img/Resolution_ratio_of_magnetic_spectrometer.png" style="zoom:60%">

- 假设一单能$E_0$的束流经过磁场后所成像的宽度为$$w$$。将宽度为$w$的狭缝置于$E_0$对应的像斑处，后置探测器，则所测能谱上的单能峰落在$E_0$处。又假设将单色束流的能量调为$\left(E_0+\mathrm{\Delta E}/2\right)$后，像斑发生$w/2$的位移，则若保持狭缝和探测器的位置不变，探头只能接受到一半的像，对应能量的计数率将减半。

- 因此如果初始束流能量是连续的均匀分布，则进入狭缝的束流的能量半高宽即为$\mathrm{\Delta E}$。在非相对论情形下有
$$\frac{\Delta{p}}{p}=\frac12\frac{\Delta{E}}{E}$$

再代入色散D的定义式，得到

$$\Delta{E}=\frac{w}{D/2}E$$

- 狭缝的宽度S对能谱的计数率和分辨率有重要影响。假设束流强度在像斑w的范围内是均匀分布，那么如果狭缝的宽度S大于w，则分辨率会降低，最大计数率不变，同时谱线出现平台；若S小于w，则分辨率不变，谱线出现平台，且最大计数率降低。可见狭缝的最佳宽度就是像斑的宽度w。实际情况中束流强度的分布十分接近均匀分布，因此这个结论依然是适用的。

- 像宽w包含两部分的贡献：像差和源的宽度。这里暂且忽略像差的影响，于是磁谱仪分辨率可以写作

$$\frac{\Delta{E}}{E}=2\frac{M_x}{D}(2\Delta{h})$$

其中$\Delta{h}$是放射源的半径。

- 在一阶成像的条件下，$\frac{D}{M_x}$与磁谱仪参数$l_1,\varepsilon_1,\phi$是正相关，与$\varepsilon_2$无关。$l_2$不是独立参数。

### 入射角的约束

- 为提高分辨率，应当尽量增大入射角$\varepsilon_1$。但在y方向$\varepsilon_1>0$作为一个凸透镜，当焦距$f=\frac{r}{\tan{\varepsilon_1}}$超出某个上下限时会有部分$\alpha$粒子击打在磁极材料上，严重降低束流传输效率。

<img src="img/Particle_trajectory_in_y_direction.jpg" style="zoom:70%">

- 入射角的最大值应恰好使得进入磁谱仪的粒子在y方向都能射出磁谱仪，如右图。最小值应使放射源位于焦点处。

- 这给出了入射角的约束条件

$$\arctan \frac{r}{l_1}\leq\varepsilon_1(\phi,l_1)\leq\arctan \bigg(\frac2{\phi}+\frac{r}{l_1}\bigg)$$

### 双聚焦

- 束流传输效率在y方向一般很低，受磁极间距的限制。如果增大磁极间距以提高效率，而没有聚焦，受限于探测器的灵敏面积，就会有部分alpha粒子打在探头之外。因此在y方向也需要实现聚焦，即双聚焦。

- 参考上图，根据透镜成像公式可得

$$\tan\varepsilon_2=\frac2{\phi}+\frac{r}{l_2}$$

将$l_2$的表达式代入，得到出射角的约束条件

$$\varepsilon_2(\phi,l_1,\varepsilon_1)=\arctan\bigg(\frac1{\phi}-\frac12\frac{\cfrac{r}{l_1}+\tan\varepsilon_1-\tan\phi}{(\cfrac{r}{l_1}+\tan\varepsilon_1)\tan\phi+1}\bigg)$$

### 二阶成像

- 在$l_2$表达式成立时，系统在xz平面达到一阶成像；此时，xz平面二阶像差的表达式为

$$\Delta{B}_2=-\kappa{r}\bigg(c_1+c_2-\frac{\gamma_1}{R_1}r-\frac{\gamma_2}{R_2}r\bigg)\Delta\varepsilon_1^2$$

- 其中

$$\kappa=\frac{1}{2}\frac{l_1^2l_2}{r^3}\bigg[1+\bigg(\frac{r}{l_1}+\tan\varepsilon_1\bigg)^2\bigg]\bigg[1+\bigg(\frac{r}{l_2}+\tan\varepsilon_2\bigg)^2\bigg]^{1/2}$$

$$c_{1,2}=\frac{r^2}{l_{1,2}^2}\frac{\cfrac{r}{l_{1,2}}+3\tan\varepsilon_{1,2}}{\bigg[1+\bigg(\cfrac{r}{l_{1,2}}+\tan\varepsilon_{1,2}\bigg)^2\bigg]^{3/2}}$$

$$\gamma_{1,2}=\frac{1}{\cos^3\varepsilon_{1,2}\bigg[1+\bigg(\cfrac{r}{l_{1,2}}+\tan\varepsilon_{1,2}\bigg)^2\bigg]^{3/2}}$$

- 因此磁场边界的曲率应当满足

$$r\bigg(\frac{\gamma_1}{R_1}+\frac{\gamma_2}{R_2}\bigg)-(c_1+c_2)=0$$

- 可见，事实上$R_1,R_2$完全由涉及一阶成像的几个参数$l_1,l_2,\varepsilon_1,\varepsilon_2,r,\phi$所确定

- yz平面对分辨率没有影响，不考虑高阶像差
