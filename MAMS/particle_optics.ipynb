# 束流光学

### 相空间

<img src="img/Beam_optics_coordinate_system.jpg" style="zoom:90%">

#### 坐标系

取束流中假想的参考粒子及其相应的参考轨道。z轴总与参考轨道相切，称为“径向”(longitunidal direction)；x轴和y轴与之正交。

#### 相空间坐标

分析力学中，相空间由3个空间坐标和3个动量坐标搭建。束流光学则将相空间坐标取为关于参考粒子的相对值。以下，下标a表示任一粒子，下标c表示参考粒子。

$$x=x_a-x_c$$

$$x'=\frac{\dot{x}_a-\dot{x}_c}{\dot{z}_c}=\frac{\dot{x}_a}{\dot{z}_c}$$

$$y=y_a-y_c$$

$$y'=\frac{\dot{y}_a-\dot{y}_c}{\dot{z}_c}=\frac{\dot{y}_a}{\dot{z}_c}$$

$$z=z_a-z_c$$

$$\delta=\frac{\dot{z}_a-\dot{z}_c}{\dot{z}_c}$$

### 矩阵法

#### 转移映射的多级展开

一般地，束流线上任意前后两点的相空间坐标q由一个转移映射(transfer map)联系，用幂级数展开得

$$M(q)=\sum_{j\leq6}R_{ij}q_j+\sum_{j\leq6}\sum_{k\leq6}T_{ijk}q_jq_k+\cdots$$

#### 矩阵光学的一些结论

- 对于两端折射率相同的系统，其R矩阵的行列式$$\det (R)=1$$

- 自由漂移（距离为d）

$$R=\begin{bmatrix}
1 & d\\
0 & 1
\end{bmatrix}$$

- 薄透镜（焦距为f）

$$R=\begin{bmatrix}
1 & 0\\
-\cfrac1f & 1
\end{bmatrix}$$

- 主平面(principle plane)

对于任意光学系统，R矩阵为

$$R=\begin{bmatrix}
A & B\\
C & D
\end{bmatrix}$$

期望通过在系统前后分别加入一段漂移距离，使得整体的R矩阵具有薄透镜的形式：（注意矩阵应“倒序相乘”，o指"object"，i指"image"）

$$\begin{bmatrix}
1 & 0\\
-\cfrac1{f_{eff}} & 1
\end{bmatrix}
=
\begin{bmatrix}
1 & p_i\\
0 & 1
\end{bmatrix}
\begin{bmatrix}
A & B\\
C & D
\end{bmatrix}
\begin{bmatrix}
1 & p_o\\
0 & 1
\end{bmatrix}$$

借助条件$\det (R)=1$，即系统两端折射率相同时，可由解出漂移距离及有效焦距

$$p_o=\frac{1-D}{C},p_i=\frac{1-A}{C},f_{eff}=-\frac{1}{C}$$

由以上两个漂移距离确定的两个平面，称为系统的主平面。

        \delta

\begin{bmatrix}

$$\begin{pmatrix}

\end{pmatrix}

\begin{bmatrix}

- xz平面

假设束流轨道不变，则相比垂直入射的情形将多感受到一部分磁场，称为wedge。在小入射角$\varepsilon$下，外侧粒子无偏转时，参考轨道有偏转，偏转角度差为$\cfrac{x\tan\varepsilon}{\rho}$；内侧粒子有偏转时，参考轨道无偏转，角度差同样为$\cfrac{x\tan\varepsilon}{\rho}$。

$$\begin{pmatrix}

\end{pmatrix}
=
\begin{bmatrix}

\end{bmatrix}
\begin{pmatrix}
	x_0\\
	x_0'\\

\end{pmatrix}$$

### 四极磁铁

<img src="img/Cross_section_of_quadrupole.png" style="zoom:80%">

四极磁铁有四个磁极，磁极表面为双曲线形状。从几何上看，四极磁铁仅有两个参数——孔径$r_0$和沿轴向的长度L。由于磁极间不包含自由电流，因此可以运用磁标势求解磁场分布。磁极表面为等磁势面，设其磁势为W，则磁极间的磁势满足二维拉普拉斯方程。解得

$$w(x,y)=\frac{2Wxy}{r_0^2}$$

磁通密度分布

$$B_x=\frac{2W}{r_0^2}y=B_T\frac{y}{r_0}$$

$$B_y=\frac{2W}{r_0^2}x=B_T\frac{x}{r_0}$$

其中$B_T$为磁极表面磁场（即磁场最大值，也可以称作特征值）$$B_T=\frac{2W}{r_0}$$

这时磁场分布就可以通过牛顿运动方程完整地解出。实际上在四极磁铁中更关心的是横向位移x, y随径向位移z的变化规律，因此需要先引入中间变量z。根据链式法则，有

$$\frac{dx}{dt}=\frac{dx}{dz}\frac{dz}{dt}$$

$$
\begin{aligned}
\frac{d^2x}{dt^2}&=\frac{d}{dt}(\frac{dx}{dz})\frac{dz}{dt}+\frac{dx}{dz}\frac{d}{dt}(\frac{dz}{dt})\\
&=\frac{d^2x}{dz^2}(\frac{dz}{dt})^2+\frac{dx}{dz}\frac{d^2z}{dt^2}
\end{aligned}
$$

这样就完成了变量替换

$$\ddot{x}=x''v_z^2+x'\ddot{z}$$

其中$\ddot{x}$是对时间t的导数，$x''$是对径向位移z的导数。在束流中可以忽略径向加速度$\ddot{z}$，因此简化为

$$\ddot{x}=x''v_z^2$$

代入牛顿运动方程得到

$$\ddot{x}=x''v_z^2=-\frac{q}{m}v_zB_y=-\frac{qv_zB_T}{mr_0}x$$

$$\ddot{y}=y''v_z^2=\frac{q}{m}v_zB_x=\frac{qv_zB_T}{mr_0}y$$

于是可以得到与中间变量z相关的运动方程

$$x''(z)+\frac{qB_T}{r_0\sqrt{2mE}}x(z)=0$$

$$y''(z)-\frac{qB_T}{r_0\sqrt{2mE}}y(z)=0$$

$$x(0)=x_0,x'(0)=x_0',y(0)=y_0,y'(0)=y_0'$$

定义角频率$$k=\sqrt{{qB_T}/{(r_0\sqrt{2mE})}}$$，结合初始条件解出特解

$$x(z)=x_0\cos(kz)+x_0'\frac{\sin(kz)}{k}$$

$$y(z)=y_0\cosh(kz)+y_0'\frac{\sinh(kz)}{k}$$

求导得

$$x'(z)=-x_0k\sin(kz)+x_0'\cos(kz)$$

$$y'(z)=y_0k\sinh(kz)+y_0'\cosh(kz)$$

其中$x_0,x_0',y_0,y_0'$为初值，即四极磁铁入口$z=0$处粒子的横向位置和速度方向。

若四极磁铁长度为L，将$z=L$代入上面的特解就可得到出口处粒子的$x,x',y,y'$。于是四极磁铁的R矩阵

$$
\begin{pmatrix}
	x\\
	x^{\prime}
\end{pmatrix}
=
\begin{bmatrix}
	\cos(kL) & \sin(kL)/k\\
	-k\sin(kL) & \cos(kL)
\end{bmatrix}
\begin{pmatrix}
	x_0\\
	x_0^{\prime}
\end{pmatrix}
$$

$$
\begin{pmatrix}
	y\\
	y^{\prime}
\end{pmatrix}
=
\begin{bmatrix}
	\cosh(kL) & \sinh(kL)/k\\
	k\sinh(kL) & \cosh(kL)
\end{bmatrix}
\begin{pmatrix}
	y_0\\
	y_0^{\prime}
\end{pmatrix}
$$
