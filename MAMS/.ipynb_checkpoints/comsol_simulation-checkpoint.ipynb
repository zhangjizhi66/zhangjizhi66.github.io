# Comsol 仿真

- 版本：Comsol Multiphysics 5.4

### 新建模型

<img src="img/Comsol新模型界面.png" style="zoom:60%">

- 打开Comsol，进入模型向导界面，选择空间维度“三维”
- 添加物理场接口，选择“磁场，无电流(mfnc)”模块，以及“带电粒子追踪(cpt)”模块
- 点击“完成”

### 添加研究

<img src="img/Comsol添加研究界面改.png" style="zoom:60%">

- 在上方菜单栏点击“研究”—“添加研究”，屏幕右侧弹出新窗口“添加研究”
- 在“添加研究”窗口下方将“磁场，无电流(mfnc)”复选框打勾，“带电粒子追踪(cpt)”复选框不打勾
- 在窗口上方点击“一般研究”—“稳态”，然后点击“添加研究”。该研究（“研究1”）将用于求解恒定磁场的分布
- 随后将“磁场，无电流(mfnc)”“带电粒子追踪(cpt)”复选框均打勾，在上方点击“一般研究”—“瞬态”，然后点击“添加研究”。该研究（“研究2”）将用于求解恒定磁场分布下粒子的运动轨迹
- 在屏幕左侧的“模型开发器”窗口点击“研究2”—“步骤1：瞬态”，在中间栏的“物理场和变量选择”栏下去掉“磁场，无电流(mfnc)”复选框的勾；再在“因变量”栏下设置“求解变量的初始值”为“物理场控制”，“不求解的变量值”依次设置为“用户控制”、“解”、“研究1，稳态”、“自动”
- 关闭“添加研究”窗口

### 设定物理条件

在“组件1”的“磁场，无电流”栏下，添加“磁通密度”，并将其“边界选择”设定为两个磁极面，随后设定磁通密度为z方向，大小为1.13T。在“带电粒子追踪”栏下，设定粒子束的各项参数。其中“相空间椭圆取样”设定为“KV”，即均匀分布，并根据表2设定“相空间椭圆尺寸”。“纵向速度”栏中选择“指定动能”，并设置“值列表”为5.445、5.5、5.555 MeV。
	最后在“网格1”的“大小1”栏中选择“超细化”，并点击“全部构建”，进行有限元子域的划分。根据计算机性能的差异可以适当选择合适的网格大小。构建完毕后，依次在“研究1”和“研究2”中点击“计算”。

### 结果显示

计算完成后，在“结果”栏中的“磁通密度模”中添加“切面”，将切面位置设定为z=0的“xy平面”，并将“表达式”设定为“mfnc.Bz”，以显示磁极中心平面上磁通密度z分量的分布情况，如图26。从中可见磁场内部的均匀性较好，但在磁极间距1.5cm的情况下边缘效应较为明显。
在“结果”栏中选择“粒子轨迹”，点击“绘制”，就可显示束流的轨迹情况，如图25。这里绘制的是单能5.5MeV的粒子轨迹。从中可以直观地看到，磁谱仪的聚焦效果是很好的；且除入口和出口处之外，束流均保持在磁场均匀性好的区域运动，尽可能地避免了边缘效应的影响。
在“结果”栏的“数据集”下新建“截面1”，将其位置设定为像平面的位置。随后新建“二维绘图组”，在其中添加“庞加莱图”。将庞加莱图的“数据”设置为“截面1”，点击“绘制”以显示束流在像平面上的分布情况，如图26。将庞加莱图的数据导出，就可使用程序计算磁谱仪的分辨率，结果为0.2037%，与理论结果以及蒙特卡洛模拟的结果相近。
